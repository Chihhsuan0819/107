```python
data_all = pd.concat([data_cycu, data_fcu, data_kuas, data_ncku, data_nthu, data_ntu, data_ntust, data_ntut, data_thu])
```

```python
#化學系化工學系會重複辨別(化工也有化)>>>學這個字還是得保留
stop_word = ['系', '研', '所', '暨']
def parser(string):
    s = ""
    i = 0
    while i < len(string) and string[i] not in stop_word:
        s += string[i]
        i += 1
    return s
```

```python
df = data_all.drop(['學校'], axis=1)
df['系所'] = df['系所'].map(parser)
```

```python
trydf = {} #先計算各個學系被歸在哪個院較多
for index, row in df.iterrows():
    if row['系所'] in trydf:
        if row['院'] in trydf[row['系所']]:
            trydf[row['系所']][row['院']] += 1
#             print(trydf[row['系所']])
        elif row['院'] not in trydf[row['系所']]:
            trydf[row['系所']][row['院']] = 1
#         print(row['系所'])
    elif row['系所'] not in trydf:
        trydf[row['系所']] = {}
        trydf[row['系所']][row['院']] = 1
#     print(row['系所'])
#     print(trydf)
    
```

```python
college = {} #再以較多的院作為專業度的依據
for key, value in trydf.items():
    pivot = 0
    kkk = ""
    for k, v in trydf[key].items():
        if v > pivot:
            kkk = k
            pivot = v
    college[key] = kkk
```

```python
if '經濟學' in college: #手動修改一些重複或有問題的部分，這部分可以再重新調整
    del college['經濟學']
if "" in college:
    del college['']
if "中國語文學" in college:
    del college['中國語文學']
if "心理學" in college:
    del college['心理學']
if "會計學學" in college:
    del college['會計學學']
if "會計學" in college:
    del college['會計學']
if "材料科學與工程學" in college:
    del college['材料科學與工程學']
if "法律學" in college:
    del college['法律學']
college['法律'] = '法律學院'
if "工業與系統工程系" in college:
    del college[' 工業與系統工程系']
```

```python
data_manager = pd.read_excel('2-1.董監經理人學經歷_金融業_2017.xlsx')
data_manager = data_manager.drop(['資料源年月', '資料源', '資料截止日',, '初任日期--董監事', '身份代碼', '獨立代碼', '會計主管', '財務主管', '重整代碼', '沿用日', '財務', '會計', '法務'], axis=1)
```

```python
data_man = data_manager.groupby(['姓名代碼'])
```

```python
df_final = pd.DataFrame()

#keyword = ['大學', '學校', '系', '專科', '學院', '商專'] 
#可以辨別是否是學歷的關鍵字，有出現這些字則高度可能是學歷

for name, group in data_man:
    dftemp = group.drop(['行次', '學經歷及目前兼任說明'], axis=1).iloc[0]
#     print(dftemp)
    dftemp['學歷'] = ''
    dftemp['經歷'] = ''
    dftemp['目前兼任'] = ''
#     print(dftemp)
    for index, row in group.iterrows(): #偵測關鍵字，含有則填入學歷的欄位，若是教授則填入經歷
        #(這裡原本想用"學"當作關鍵字，結果發現職位叫"學習型組織主管"，只好改用複雜的方式處理
#         print(row['行次'])
        
        if '大學' in row['學經歷及目前兼任說明']:
            if '教授' in row['學經歷及目前兼任說明']:
                dftemp['經歷'] += row['學經歷及目前兼任說明'] + '、'
            else:
                dftemp['學歷'] += row['學經歷及目前兼任說明'] 
        elif '學院' in row['學經歷及目前兼任說明']:
            if '教授' in row['學經歷及目前兼任說明']:
                dftemp['經歷'] += row['學經歷及目前兼任說明'] + '、'
            else:
                dftemp['學歷'] += row['學經歷及目前兼任說明']
        elif '專科' in row['學經歷及目前兼任說明']:
            if '教授' in row['學經歷及目前兼任說明']:
                dftemp['經歷'] += row['學經歷及目前兼任說明'] + '、'
            else:
                dftemp['學歷'] += row['學經歷及目前兼任說明']
        elif '學校' in row['學經歷及目前兼任說明']:
            if '教授' in row['學經歷及目前兼任說明']:
                dftemp['經歷'] += row['學經歷及目前兼任說明'] + '、'
            else:
                dftemp['學歷'] += row['學經歷及目前兼任說明']
        elif '商專' in row['學經歷及目前兼任說明']:
            if '教授' in row['學經歷及目前兼任說明']:
                dftemp['經歷'] += row['學經歷及目前兼任說明'] + '、'
            else:
                dftemp['學歷'] += row['學經歷及目前兼任說明']
                
        elif row['行次'][0] == 'A': #其他的部分則按照 A=經歷，B=目前兼任的原則填入
            dftemp['經歷'] += row['學經歷及目前兼任說明'] + '、'
        elif row['行次'][0]  == 'B':
            dftemp['目前兼任'] += row['學經歷及目前兼任說明'] + '、'
    
    df_final = df_final.append(dftemp)
```

```python
def whichcollege(string): #寫入專業度的函數
    j = 0
    while  j < len(list(college.items())) and list(college.items())[j][0] not in string:
        j = j + 1
    if j < len(list(college.items())):
        return list(college.items())[j][1]
    elif j >= len(list(college.items())):
        return('無法辨別')
```
